```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
#Create dataset with each row representing one order sequence 
dataset=[["Bread","Milk","Beer"],["Bread","Diapers","Eggs"],["Milk","Diapers","Beer","Cola"],["Bread","Milk","Diapers","Beer"],["Bread","Milk","Cola"]]
dataset
```

```python
te=TransactionEncoder()
#Read all unique items
te=te.fit(dataset)   
#creates a matrix of orders
te_ary=te.transform(dataset) 
print(te.columns_)
print(te_ary)
```

```python
#Binary encoded orders
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
#filter items having sufficient support
frequent_itemsets=apriori(df,min_support=0.4,use_colnames=True)
frequent_itemsets
```

```python
#Evaluate items with good support and confidence = 0.7
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7)
rules
```
