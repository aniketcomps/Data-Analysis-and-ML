### Apriori Algorithm

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
dataset=[["Bread","Milk","Beer"],["Bread","Diapers","Eggs"],["Milk","Diapers","Beer","Cola"],["Bread","Milk","Diapers","Beer"],["Bread","Milk","Cola"]]
dataset
```

```python
te=TransactionEncoder()
te=te.fit(dataset)   #Read all unique items
te_ary=te.transform(dataset)  # creates a matrix
print(te.columns_)
print(te_ary)
```

```python
df=pd.DataFrame(te_ary,columns=te.columns_)
#df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
#filter products
frequent_itemsets=apriori(df,min_support=0.4,use_colnames=True)
frequent_itemsets
```

```python
#Evaluate items with good support and confidence = 0.5
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.5)
rules
```
